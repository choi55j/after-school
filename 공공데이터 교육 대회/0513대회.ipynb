```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
data = pd.read_csv("장학1.csv", encoding='cp949',low_memory=False)
```

```python
data['교내장학금 소계(원)'] = pd.to_numeric(data['교내장학금 소계(원)'], errors='coerce').fillna(0)
data['교내장학금 성적우수장학금'] = pd.to_numeric(data['교내장학금 성적우수장학금'], errors='coerce').fillna(0)
data['교내장학금 저소득층장학금'] = pd.to_numeric(data['교내장학금 저소득층장학금'], errors='coerce').fillna(0)
data['교내장학금 근로장학금'] = pd.to_numeric(data['교내장학금 근로장학금'], errors='coerce').fillna(0)
data['교내장학금 재난장학금'] = pd.to_numeric(data['교내장학금 재난장학금'], errors='coerce').fillna(0)
data['교내장학금 교직원장학금'] = pd.to_numeric(data['교내장학금 교직원장학금'], errors='coerce').fillna(0)
data['교내장학금 기타장학금'] = pd.to_numeric(data['교내장학금 기타장학금'], errors='coerce').fillna(0)
교내장학금 = data[['학교명','교내장학금 성적우수장학금','교내장학금 저소득층장학금','교내장학금 근로장학금','교내장학금 재난장학금','교내장학금 교직원장학금','교내장학금 기타장학금']]
교내장학금.set_index("학교명", inplace = True)

교내장학금.sort_values(by='교내장학금 성적우수장학금',ascending=False).head(10).plot(kind='bar')
```
