## 교외장학금 

```python
data['교외장학금 소계(원)'] = pd.to_numeric(data['교외장학금 소계(원)'], errors='coerce').fillna(0).sort_values()
data['교외장학금 국가'] = pd.to_numeric(data['교외장학금 국가'], errors='coerce').fillna(0).sort_values()
data['교외장학금 지방자치단체'] = pd.to_numeric(data['교외장학금 지방자치단체'], errors='coerce').fillna(0).sort_values()
data['교외장학금 사설및기타'] = pd.to_numeric(data['교외장학금 사설및기타'], errors='coerce').fillna(0).sort_values()
교외장학금 = data[['학교명','교외장학금 소계(원)','교외장학금 국가','교외장학금 지방자치단체','교외장학금 사설및기타']]
교외장학금.set_index("학교명", inplace=True)

교외장학금.sort_values(by='교외장학금 소계(원)', ascending=False).head(10).plot(kind="bar")
```

```python
data['교내장학금 소계(원)'] = pd.to_numeric(data['교내장학금 소계(원)'], errors='coerce').fillna(0)
data['교내장학금 성적우수장학금'] = pd.to_numeric(data['교내장학금 성적우수장학금'], errors='coerce').fillna(0)
data['교내장학금 저소득층장학금'] = pd.to_numeric(data['교내장학금 저소득층장학금'], errors='coerce').fillna(0)
data['교내장학금 근로장학금'] = pd.to_numeric(data['교내장학금 근로장학금'], errors='coerce').fillna(0)
data['교내장학금 재난장학금'] = pd.to_numeric(data['교내장학금 재난장학금'], errors='coerce').fillna(0)
data['교내장학금 교직원장학금'] = pd.to_numeric(data['교내장학금 교직원장학금'], errors='coerce').fillna(0)
data['교내장학금 기타장학금'] = pd.to_numeric(data['교내장학금 기타장학금'], errors='coerce').fillna(0)
교내장학금 = data[['학교명','교내장학금 소계(원)','교내장학금 성적우수장학금','교내장학금 저소득층장학금','교내장학금 근로장학금','교내장학금 재난장학금','교내장학금 교직원장학금','교내장학금 기타장학금']]

교내장학금.sort_values(by='교내장학금 소계(원)',ascending=False).head(10).plot(kind='bar')
```
