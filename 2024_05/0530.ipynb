```python
import pandas as pd
from sklearn.datasets import load_iris
# sklearn 패키지에서 붓꽃(iris) 데이터를 가져온다.
iris = load_iris()
# 붓꽃의 특성을 가져온다.
iris_data =iris.data
# 붓꽃의 품종을 가져온다.
iris_target = iris.target

iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names)
# DataFrame을 생성한 이후에 target 값들을 'label'이라는 새로운 열을 만들어 넣는다. 
iris_df['label'] = iris_target

iris_df.tail(5)
```

## 데이터 전처리 -> 데이터 세트 분리 -> 모델 학습 및 검증 평가 -> 예측 수행 -> 예측 평가

### 데이터 전처리 예시

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# 범주형 데이터를 가진 배열 생성
categories = np.array(['사과', '바나나', '오렌지', '사과', '바나나'])

# OneHotEncoder 객체 생성
encoder = OneHotEncoder(sparse=False)

# 범주형 데이터를 원-핫 인코딩
one_hot_encoded = encoder.fit_transform(categories.reshape(-1, 1))

print("인코딩된 데이터:")
print(one_hot_encoded)
```

```python
import pandas as pd

# 간단한 데이터프레임 생성
data = pd.DataFrame({'과일': ['사과', '바나나', '오렌지', '사과', '바나나']})

# 원-핫 인코딩 수행
one_hot_encoded = pd.get_dummies(data)

print("인코딩된 데이터:")
print(one_hot_encoded)
```
