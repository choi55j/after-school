```python
arr = [7, 9, 8, 6, 5, 4, 3, 2, 1, 10]

def swap(arr, i, j):
    arr[i], arr[j] = arr[j], arr[i]

def sort(arr):
    length = len(arr)
    for i in range(length):
        for j in range(length - 1):
            if arr[j] > arr[j+1]: # 내림차순 정렬을 위해 부등호 방향 수정
                swap(arr, j, j+1)

sort(arr)
print(arr)
```

```python
def bubble_sort(arr1):
    for i in range(len(arr1)):
        for j in range(len(arr1) - i, - 1):
            if arr1[j] > arr1[j+1]:
                arr1[j], arr1[j+1] = swap(arr1[j+1], arr1[j])
    return arr1

list1 = [1,3,6,4,5,7,9,8,10,2]
sorted = bubble_sort(list1)
print(sorted)
```
