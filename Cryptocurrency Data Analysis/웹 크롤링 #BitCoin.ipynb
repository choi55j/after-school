```python
import requests
from bs4 import BeautifulSoup

# CNN 뉴스 페이지 URL (예: 최신 뉴스 페이지 URL)
url = 'https://edition.cnn.com/world'

# HTTP GET 요청
response = requests.get(url)
response.raise_for_status()  # 요청이 실패할 경우 예외를 발생시킵니다.

# HTML 파싱
soup = BeautifulSoup(response.text, 'html.parser')

# 뉴스 기사 제목과 링크 추출
for article in soup.find_all('h3', class_='cd__headline'):
    title = article.get_text()
    link = article.find('a')['href']
    if not link.startswith('http'):
        link = 'https://edition.cnn.com' + link
    print(f'Title: {title}')
    print(f'Link: {link}')
    print()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# 웹 드라이버 설정
chrome_options = Options()
chrome_options.add_argument("--headless")  # 헤드리스 모드로 실행 (GUI 없이)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# CNN 뉴스 페이지 URL
url = 'https://edition.cnn.com/world'
driver.get(url)

# 페이지 로딩 대기 (동적 콘텐츠 로딩을 기다리기 위해)
time.sleep(5)

# 뉴스 기사 제목과 링크 추출
articles = driver.find_elements(By.CSS_SELECTOR, 'h3.cd__headline')
for article in articles:
    title = article.text
    link = article.find_element(By.TAG_NAME, 'a').get_attribute('href')
    print(f'Title: {title}')
    print(f'Link: {link}')
    print()

# 드라이버 종료
driver.quit()
```
