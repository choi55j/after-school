```python
큐 = [0,0,0,0,0,0]

앞쪽 = -1
뒤쪽 = -1

while True:
    print('(1)삽입 (2)삭제 (3)출력 (4)종료')
    선택 = int(input())
    if 선택 == 1:
        앞쪽 +=1
        if 앞쪽==6:
            앞쪽-=1
            print('더 이상 추가가 되지 않습니다.')
        else:
            큐[앞쪽] = int(input('값을 입력하세요'))
    elif 선택==2:
        뒤쪽 +=1
        if 앞쪽 == 뒤쪽:
            print('초기화 할게???')
            큐 = [0,0,0,0,0,0]
            뒤쪽=-1
            앞쪽=-1
        else:
            print(f'{큐[뒤쪽]} 값을 삭제 합니다.')
    elif 선택==3:
        for i in range(뒤쪽+1,앞쪽+1):
            print(큐[i],end='')
        print()
    else:
        break
        print('종료 되었습니다')
```
