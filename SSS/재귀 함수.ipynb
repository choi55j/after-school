## 팩토리얼 구하는 프로그램

```python
# 팩토리얼(!)
# 반복문을 이용한 팩토리얼 구하기
# 변수 팩토리얼 값을 담을 변수, 숫자 변수

n = int(input())
p = 1
for i in range(n,0,-1):
    p *=i
    print(f'현재 i : {i}, 현재 팩토리얼 : {p}')
print(p)
```

```python
def 팩토리얼(n):
    if n==1:
        print(f'현재i : {n}, 팩토리얼을 종료합니다')
        return 1
    print(f'현제 숫자 {n}, 현재 팩토리얼 {n-1}를 호출합니다')
    return n*팩토리얼(n-1)


숫자 = int(input('숫자? '))
팩토리얼(숫자)
```

```python
def pack(n):
    return n*n-1

print(pack)

n = int(input())
pack(n)
```
